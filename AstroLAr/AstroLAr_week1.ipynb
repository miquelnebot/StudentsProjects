```python
# Week 1: Energy Spectra of Astrophysical Neutrino Sources
 ##ðŸ“Š Produce a comparative plot of typical neutrino energy spectra for:
#Supernova (SN)
#Solar
#Atmospheric
#Diffuse Supernova Neutrino Background (DSNB)
#Gamma Ray Bursts (GRBs)
```

```python
## Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Define mock spectra (log scale energy in MeV)
energies = np.logspace(-1, 2, 500)  # 0.1 MeV to 100 MeV
```

```python
# Example spectra (simplified placeholders â€“ to be replaced with literature values)
def sn_spectrum(E):
    return E**3 * np.exp(-E/10)  # thermal

def solar_spectrum(E):
    return 1e-1 * np.exp(-E/1)  # lower energy

def atm_spectrum(E):
    return 1e-4 * E**-2.7  # power-law

def dsnb_spectrum(E):
    return 1e-3 * E**3 * np.exp(-E/6)  # thermal relic

def grb_spectrum(E):
    return 1e-6 * E**-2  # harder spectrum
```

```python
# Normalised spectra
plt.figure(figsize=(10,6))
plt.loglog(energies, sn_spectrum(energies), label='Core-Collapse SN')
plt.loglog(energies, solar_spectrum(energies), label='Solar Neutrinos')
plt.loglog(energies, atm_spectrum(energies), label='Atmospheric')
plt.loglog(energies, dsnb_spectrum(energies), label='DSNB')
plt.loglog(energies, grb_spectrum(energies), label='GRBs')

plt.xlabel('Neutrino Energy [MeV]')
plt.ylabel('Arbitrary Units (Flux Shape)')
plt.title('Neutrino Energy Spectra from Astrophysical Sources')
plt.legend()
plt.grid(True, which='both', ls='--')
plt.tight_layout()
plt.savefig('../plots/energy_spectra.png', dpi=300)
plt.show()
```
